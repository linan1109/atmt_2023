```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch_size = ['1','8','16','32','64','128']
batch_size_bleu = [17.1,17.5,11.5,11.7,9.9,13.2]

plt.plot(batch_size, batch_size_bleu)
plt.scatter(batch_size, batch_size_bleu, marker='x', s=100)
for i in range(len(batch_size)):
    plt.annotate(batch_size_bleu[i], xy=(batch_size[i], batch_size_bleu[i]), xytext=(batch_size[i], batch_size_bleu[i]), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Batch Size', fontsize=14)
plt.ylabel('BLEU', fontsize=14)
plt.title('BLEU in different Batch Size', fontsize=16)
```

```python
dropout = ['0', '0.3', '0.5', '0.7']
dropout_bleu = [9.8,15.5,8.2,0.5]

plt.figure()
plt.plot(dropout, dropout_bleu)
plt.scatter(dropout, dropout_bleu, marker='x', s=100)
for i in range(len(dropout)):
    plt.annotate(dropout_bleu[i], xy=(dropout[i], dropout_bleu[i]), xytext=(dropout[i], dropout_bleu[i]), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Dropout', fontsize=14)
plt.ylabel('BLEU', fontsize=14)
plt.title('BLEU in different Dropout', fontsize=16)
```

```python
lr = ['0.0001', '0.0005', '0.001', '0.005', '0.01']
lr_bleu = [15.8,14.7,9.3,0.2,0.1]

plt.figure()
plt.plot(lr, lr_bleu)
plt.scatter(lr, lr_bleu, marker='x', s=100)
for i in range(len(lr)):
    plt.annotate(lr_bleu[i], xy=(lr[i], lr_bleu[i]), xytext=(lr[i], lr_bleu[i]), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Learning Rate', fontsize=14)
plt.ylabel('BLEU', fontsize=14)
plt.title('BLEU in different Learning Rate', fontsize=16)
```

```python
model_num_layers = ['1', '2', '3']
model_num_layers_bleu = [17.1,17.9,13.5]

plt.figure()
plt.plot(model_num_layers, model_num_layers_bleu)
plt.scatter(model_num_layers, model_num_layers_bleu, marker='x', s=100)
for i in range(len(model_num_layers)):
    plt.annotate(model_num_layers_bleu[i], xy=(model_num_layers[i], model_num_layers_bleu[i]), xytext=(model_num_layers[i], model_num_layers_bleu[i]), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Number of Layers', fontsize=14)
plt.ylabel('BLEU', fontsize=14)
plt.title('BLEU in different Number of Layers', fontsize=16)
```
